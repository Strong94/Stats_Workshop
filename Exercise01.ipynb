```python
import math;
import matplotlib as mplt;
from matplotlib import pyplot as plt;
import numpy as np;
```

Assign omegam arbitrary number so we can identify s, might change s to be a function of omegam.

```python
def s( omegam ):
  return( ((1-omegam)/omegam)**(1/3))
```

Next define eta - remember multiplication signs! Doesn't fill them in like mathematica

```python
def eta( a, omegam ):
  pref=2*(s(omegam)**3+1);
  mainb=1/a**4-0.1540*s(omegam)/(a**3)+0.4304*(s(omegam)**2)/(a**2)+0.19097*(s(omegam)**3)/a+0.066941*s(omegam)**4;
  return pref*(mainb**(-1/8));
```

Next define D_L

```python
def dstarl( z, omegam ):
  return 3000*(1+z)*(eta(1,omegam)-eta(1/(1+z),omegam))
```

Finally define distance modulus

```python
def mumod( z, h, omegam ):
  return 25 - 5*math.log(h,10)+5*math.log(dstarl(z,omegam),10)
```

```python
points = 10000 #number of points on graph, range needs integer
zmin,zmax=0,2
zlist=list(map(lambda z: float(zmax-zmin)*z/points, range(points+1)))
zlist.remove(zlist[0])#want greater than and not equal to zero
zlist.remove(zlist[-1])#want less than and not equal to 2
mulist1=list(map(lambda y: mumod(y,0.7,0.2),zlist))#map gives generators, plot wants list
mulist4=list(map(lambda y: mumod(y,0.7,0.5),zlist))#map gives generators, plot wants list
mulist2=list(map(lambda y: mumod(y,0.7,0.3),zlist))#map gives generators, plot wants list
mulist3=list(map(lambda y: mumod(y,0.7,0.4),zlist))#map gives generators, plot wants list
plt.plot(zlist,mulist1)
plt.plot(zlist,mulist2)
plt.plot(zlist,mulist3)
plt.plot(zlist,mulist4)
plt.show()
```
